Q1. Stable top-K Selection.

```python
items=[("A",80),("B",90),("C",85),("D",95)]
k=2

def top_K(items,k):
    items.sort(key=lambda x:x[1], reverse=True)
    return items[:k]
```

Q3. Missing Number (0 to n)

```python
nums=[3,0,1]

def missing_number(nums):
    n=len(nums)
    expected_sum= n*(n+1)//2
    actual_sum=sum(nums)
    return expected_sum-actual_sum
```

Q5. Moving average of last w elements.

each index shows average of last w elements up to that position.

```python
stream = [1,2,3,4,5]
w=3
def moving_average(stream,w):
    window=[]
    for num in stream:
        window.append(num)
        if len(window)>w:
            window.pop(0)
            average=sum(window)/len(window)
            print(average)
return moving_average
```
